# Breath-first Search

```
class STATE:
    white = 0
    gray = 1
    black = 2
    
def bfs(g, s):
  '''node by node bfs using queue''' 
  v = len(g)
  state = [False] * v
  
  # allocate space for the predecessor list and colors 
  pi = [None] * v
  state[s] = True # make the state of the visiting node
  dist = [0] * v
  
  q, orders = [s], [s]
  while q:
    u = q.pop(0)
    
    print(u, ' out, ', end = ' ')
    for v in g[u]:
      if not state[v]:
        state[v] = True
        pi[v] = u # set the predecessor
        dist[v] = dist[u] + 1
        q.append(v)
        orders.append(v)
        print(v, ' in', end = ' ')
  return orders, pi, dist
        
```

```
def get_path(s, t, pi):
  '''iterative'''
  p = t
  path = []
  while p != s:
    path.append(p)
    p = pi[p]
  path.append(s)
  return path[::-1]
    
```

```
def bfs(g, s):
  '''simplified bfs'''
  v = len(g)
  colors = [STATE.white] * v
  
  q, orders = [s], [s]
  complete_orders = []
  colors[s] = STATE.gray # make the state of the visiting node
  while q:
    u = q.pop(0)
    
    for v in g[u]:
      if colors[v] == STATE.white:
        colors[v] = STATE.gray
        q.append(v)
        orders.append(v)
    # complete 
    colors[u] = STATE.black
    complete_orders.append(u)
  return orders, complete_orders
```

#### level by level bfs

```
def bfs_level(g, s):
  '''level by level bfs'''
  v = len(g)
  state = [False] * v
  
  orders = []
  lst = [s]
  state[s] = True
  d = 0 # track distance
  while lst:
    print('distance ', d, ': ', lst)
    tmp_lst = []
    for u in lst:
      orders.append(u)
      for v in g[u]:
        if not state[v]:
          state[v] = True
          tmp_lst.append(v)    
    lst = tmp_lst
    d += 1
  return orders
          
```

```
print(bfs_level(al, 0))
```

# Depth-first Search

```
def dfs(g, s, colors, orders, complete_orders):
  colors[s] = STATE.gray
  orders.append(s)
  for v in g[s]:
    if colors[v] == STATE.white:
      dfs(g, v, colors, orders, complete_orders)
  # complete
  colors[s] = STATE.black # this is not necessary in the code, just to help track the state
  complete_orders.append(s)
  return
```

```
def dftIter(g, s):
  '''not preserving the same discovery ordering'''
  n = len(g)
  orders = []
  colors = [STATE.white] * n
  stack = [s]

  orders.append(s) # track gray order
  colors[s] = STATE.gray
        
  while stack:
    u = stack.pop()
    
    for v in g[u]:
      if colors[v] == STATE.white:
        colors[v] = STATE.gray
        stack.append(v)
        orders.append(v) # track gray order
    
  return orders
```

```
def dfsIter(g, s):
  '''iterative dfs'''
  v = len(g)
  orders, complete_orders = [], []
  colors = [STATE.white] * v
  stack = [s]

  orders.append(s) # track gray order
  colors[s] = STATE.gray
        
  while stack:
    u = stack[-1]
    bAdj = False
    for v in g[u]:
      if colors[v] == STATE.white:
        colors[v] = STATE.gray
        stack.append(v)
        orders.append(v) # track gray order
        bAdj = True
        break
     
    if not bAdj: # if no adjacent is found, pop out
      # complete
      colors[u] = STATE.black # this is not necessary in the code, just to help track the state
      complete_orders.append(u)
      stack.pop()
    
  return orders, complete_orders
    

    
        
```
