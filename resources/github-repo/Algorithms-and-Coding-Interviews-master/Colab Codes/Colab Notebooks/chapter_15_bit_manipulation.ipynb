```

'''built-in functions'''
print(bin(1), bin(-1))
print(bin(126), bin(-126))
a = bin(88)
a1 = bin(-88)
print(a)
print(a1)
b = int('01011000', 2)
c = int('88', 10)
print(b, c)

d = chr(88)
print(d)

e = ord('a')
print(e)

empty_bytes = bytes(4)
print(type(empty_bytes))
print(empty_bytes)

# Cast bytes to bytearray
mutable_bytes = bytearray(b'\x00\x0F')

# Bytearray allows modification
mutable_bytes[0] = 255
mutable_bytes.append(255)
print(mutable_bytes)

# Cast bytearray back to bytes
immutable_bytes = bytes(mutable_bytes)
print(immutable_bytes)

# Some bytes to play with
byte1 = int('11110000', 2)  # 240
byte2 = int('00001111', 2)  # 15
byte3 = int('01010101', 2)  # 85

# Ones Complement (Flip the bits)
print(~byte1)

# AND
print(byte1 & byte2)

# OR
print(byte1 | byte2)

# XOR
print(byte1 ^ byte3)

# Shifting right will lose the right-most bit
print(byte2 >> 3)

# Shifting left will add a 0 bit on the right side
print(byte2 << 1)

# See if a single bit is set
bit_mask = int('00000001', 2)  # Bit 1
print(bit_mask & byte1)  # Is bit set in byte1?
print(bit_mask & byte2)  # Is bit set in byte2?

'''Combined bit manipulation'''
def get_bit(val, i):
    mask = 1 << i
    if val & mask:
        return 1
    return 0

# i i-1 i-2 ... 2 1 0, keep these positions
def clear_bits_left_right(val, i):
    print('val', bin(val))
    mask = (1 << i) -1
    print('mask', bin(mask))
    return bin(val & (mask))
# i i-1 i-2 ... 2 1 0, erase these positions
def clear_bits_right_left(val, i):
    print('val', bin(val))
    mask = (1 << i) -1
    print('mask', bin(~mask))
    return bin(val & (~mask))


print(get_bit(5,1))

print(clear_bits_left_right(int('11111111',2), 5))
print(clear_bits_right_left(int('11111111',2), 5))

'''applications'''
def twos_complement(val, bits):
    # first flip implemented with xor of val with all 1's
    flip_val = val ^ (1 << bits - 1)
    #flip_val = ~val we only give 3 bits
    return bin(flip_val + 1)


def twos_complement2(val, bits):
    zeroFound = False
    ans = 0
    mask = 1
    for i in range(bits):
        b = (val & (mask)) # get ith bit
        print(b)
        b = not b # flipped
        if not zeroFound:
            if not b: # found zero, flip to one, else flip to zero: no operation needed
                print('found')
                ans = ans | (mask) # set ith bit
                zeroFound = True
        else:
            if b:
                ans = ans | (mask)
        mask = mask << 1  # change mask to the next bit
    return bin(ans)

def twos_complement_result(x):
    ans1 = -x
    ans2 = ~x + 1
    print(ans1, ans2)
    print(bin(ans1), bin(ans2))
    return ans1

def strip_last_set_bit(val):
    print(bin(val))
    return bin(val & (val - 1))

def get_lowest_set_bit(val):
    return val ^ (val & (val -1))
    return (val & (-val))

print(twos_complement(5, 8))
print(twos_complement2(5, 8))
print(strip_last_set_bit(5))
print(get_lowest_set_bit(8))

twos_complement_result(8)
```
