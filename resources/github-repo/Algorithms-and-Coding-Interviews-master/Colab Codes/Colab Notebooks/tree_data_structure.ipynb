### N-aray Tree

#### Define Tree Node

```
class NaryNode:
  '''Define a n-ary node'''
  def __init__(self, n, val):
    self.children = [None] * n
    self.val = val
    
```

### Binary Tree

#### Define Tree Node

```
class BinaryNode:
  '''Define a classical binary tree node'''
  def __init__(self, val):
    self.left = None
    self.right = None
    self.val = val
```

#### Define a Tree and implement operations
      1
    /   \  
   2      3
 /   \      \
4     5      6 

```
root = BinaryNode(1)
left = BinaryNode(2)
right = BinaryNode(3)
root.left = left
root.right = right
left.left = BinaryNode(4)
left.right = BinaryNode(5)
right.right = BinaryNode(6)
```

```
def constructTree(a, idx):
  '''construct a binary tree recursively from input array a'''
  if idx >= len(a):
    return None
  node = BinaryNode(a[idx])
  node.left = constructTree(a, 2*idx + 1)
  node.right = constructTree(a, 2*idx + 2)
  return node
```

```
nums = [1, 2, 3, 4, 5, None, 6]
root = constructTree(nums, 0)
```

#### To show the nodes at each level, we use LevelOrder function to print out the tree:

```
def LevelOrder(root):
    q = [root]
    while q:
        new_q = []
        for n in q:
            if n is not None:
                print(n.val, end=',')
            if n.left:
                new_q.append(n.left)
            if n.right:
                new_q.append(n.right)
        q = new_q
        print('\n')
```
