## Fixed sliding window

```
def fixedSlideWindow(A, k):
    n = len(A)
    if k >= n:
        return sum(A)
    # compute the first window
    acc = sum(A[:k])
    ans = acc
    # slide the window
    for i in range(n-k): # i is the start point of the window
        j = i + k # j is the end point of the window
        acc = acc - A[i] + A[j]
        ans = max(ans, acc)
    return ans
```

```
A =[8,5,10,7,9,4,15,12,90,13]
fixedSlideWindow(A,3)
```

## Plot the prefix sum

```
nums = [2,3,1,2,4,3]
s = 7

prefixSum = [0]
for n in nums:
    prefixSum.append(prefixSum[-1]+n)
```
